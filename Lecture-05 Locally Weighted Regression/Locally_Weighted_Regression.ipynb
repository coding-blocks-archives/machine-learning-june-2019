```python
def getWMatrix(X,X_Test,tau):
    
    m = X.shape[0]
    W = np.zeros((m,m))
    
    for i in range(m):
        val = np.exp(-np.dot((X[i]-X_Test).T,(X[i]-X_Test))/(2*tau*tau))
        W[i,i] = val #np.exp(-np.dot((X[i]-X_Test).T,(X[i]-X_Test))/(2*tau*tau))
        
    return W
    
```
