### Dataset Preparation 
Generating a Random Dataset (X-One Feature) and Y-Predicted value

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.randn(100)
```

```python
e = np.random.randn(100)
Y = 2*X + 5 + 0.3*e
```

```python
plt.scatter(X,Y)
plt.xlabel("House Area")
plt.ylabel("Price")
plt.show()
```

```python
def grad(X,Y,theta):
    
    m = X.shape[0]
    grad = np.zeros((2,))
    
    for i in range(m):
        yi = theta[1]*X[i] + theta[0]
        grad[0] += 2*(yi - Y[i])
        grad[1] += 2*(yi - Y[i])*X[i]
    
    return grad
    
```

```python
def total_error(X,Y,theta):
    m = X.shape[0]
    e = 0 
    for i in range(m):
        yi = theta[1]*X[i] + theta[0]
        e += (yi-Y[i])**2
    
    return e
```

```python

def gradientDescent(X,Y,learning_rate=0.001,max_steps=80):
    theta = np.zeros((2,))
    error_list = [] 
    theta_list = []
    
    for i in range(max_steps):
        #update theta
        g = grad(X,Y,theta) 
        theta[0] = theta[0] - learning_rate*g[0]
        theta[1] = theta[1] - learning_rate*g[1]
        error = total_error(X,Y,theta)
        error_list.append(error)
        theta_list.append(theta)
        
    return theta,theta_list,error_list
    
```

```python
theta,theta_list,error_list = gradientDescent(X,Y)
```

```python
plt.plot(error_list)
plt.show()
```

```python
def predict(X_Test,theta):
    return theta[0] + theta[1]*X_Test
```

```python
X_Test = np.arange(-3,5)
```

```python
print(X_Test)
Y_Pred = predict(X_Test,theta)
print(Y_Pred)
```
